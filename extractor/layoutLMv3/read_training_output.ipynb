```python
path = os.path.join(job_dir, 'loss_history.xlsx')
training_loss = pd.ExcelFile(path)
print(training_loss.sheet_names)

training_loss_df = training_loss.parse(training_loss.sheet_names[0])

print(training_loss_df)
# plotting loss

plt.figure(figsize=(10, 6))
plt.scatter(training_loss_df['Epoch'], training_loss_df['Train Loss'],  color='blue', alpha=0.5, label='Train Loss')
plt.scatter(training_loss_df['Epoch'], training_loss_df['Validation Loss'], color='red', alpha=0.5, label='Validation Loss')
plt.title('Epoch vs Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend() 
plt.grid(True)
plt.show()
```

```python
path = os.path.join(job_dir, 'test_classification_report.xlsx')
test_metric = pd.ExcelFile(path)
print(test_metric.sheet_names)

test_metric_df = test_metric.parse(test_metric.sheet_names[0])

test_metric_df
```

```python
# Sort the DataFrame by F1 Score for better visualization
test_metric_df_sorted = test_metric_df.sort_values(by='f1-score', ascending=False)

# Plotting
plt.figure(figsize=(12, 8))
plt.barh(test_metric_df_sorted['Field'], test_metric_df_sorted['Precision'], color='skyblue', label='Precision')
plt.barh(test_metric_df_sorted['Field'], test_metric_df_sorted['Recall'], color='salmon', label='Recall')
plt.barh(test_metric_df_sorted['Field'], test_metric_df_sorted['F1 Score'], color='lightgreen', label='F1 Score')

plt.xlabel('Score')
plt.title('Performance Metrics by Field')
plt.legend()
plt.gca().invert_yaxis()  # Invert y-axis to display the highest score on top
plt.show()
```
